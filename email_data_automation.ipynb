# Process Automation using Python


 - This program will:
 - Access Google Drive > Download a file > Extract information > Send by e-mail
 - Dataset: https://www.kaggle.com/gregorut/videogamesales


```python
import pyautogui as pg
import pyperclip as pc
import time
import pandas as pd

#It needs a pause between a command and the next one (in seconds)
pg.PAUSE = 1

#Open a new tab
pg.hotkey("ctrl","t")

#Copy and paste the link 
pc.copy("https://docs.google.com/spreadsheets/d/1sQ7S8Cb06WQkJZNQBj28JIy1Jgjem1Au/edit?usp=sharing&ouid=104384067204412215105&rtpof=true&sd=true")
pg.hotkey("ctrl","v")
pg.press("enter")

#Time to load  the information(in seconds)
time.sleep(10)

#Download the data
pg.click(x=64, y=104)
pg.moveTo(x=156, y=315)

pg.move(200, 0)
pg.click(clicks=2)
```

```python
#Change to your directory
df = pd.read_excel(r"C:\Users\Fernanda\Downloads\vgsales.xlsx", index_col=0)

display(df)
```

```python
#Table info
print(df.info())
```

```python
#Treating the dataframe

#Converting year to integer
df[['Year']] = df[['Year']].fillna(value=0)
df = df.astype({"Year": "int64"})

#Dropping null values from Publisher column
df = df.dropna(how="any", axis =0)

#Table info
print(df.info())
```

```python
#Groupíng and defining variables

df_group = df.groupby(['Platform']).sum()
df_group = df_group.drop("Year", axis=1)

#Platform and value of max EU Sales
df_eu_sales = df_group.sort_values('EU_Sales', ascending=False)[:3]
eu_sales = df_eu_sales.iloc[0].values[1]
eu_plat  = df_eu_sales.iloc[0].name

#Platform and value of max Global Sales
df_global_sales = df_group.sort_values('Global_Sales', ascending=False)[:3]
global_sales = df_global_sales.iloc[0].values[4]
global_plat  = df_global_sales.iloc[0].name
```

```python
#Send email using g-mail

gmail = "https://mail.google.com/mail/u/0/#inbox"

pg.hotkey("ctrl","t")
pc.copy(gmail)
pg.hotkey("ctrl","v")
pg.press("enter")

time.sleep(5)

pg.click(x=36, y=170, clicks = 1)

time.sleep(6)

#Recipient's email address
pg.write("xxxxx@gmail.com")
pg.press("tab")
pg.press("tab") 

#Email subject
pg.write ("Daily Sales")
pg.press("tab")

text=f"""
Good morning,

Considering EU, the platform with the highest sales amount is {eu_plat}, selling a total of {eu_sales:,.2f}.
Talking about Global sales, {global_plat} is the number one platform representing a total of {global_sales:,.2f}.

Regards

"""
pc.copy(text)
pg.hotkey("ctrl","v")
pg.hotkey("ctrl","enter")
```
